```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
# 提取数据集
iris = load_iris()
y = iris.target
X = iris.data
X.shape
```

```python
import pandas as pd
pd.DataFrame(X)
```

```python
# 调用PCA 
pca = PCA(n_components = 2) # 实例化
X_dr = pca.fit_transform(X)

# 可简写成下面的形式
# X_dr = PCA(2).fit_transform(X)
X_dr
```

#### 这里的X_dr虽然是特征，但是其依然可以利用标签的来进行筛选相应的数值？目前这个还不是特别的懂

```python
X_dr[y == 0,1]
```

```python
plt.figure()
plt.scatter(X_dr[y == 0,1],X_dr[y == 0,0],c = 'red',label = iris.target_names[0])
plt.scatter(X_dr[y == 1,1],X_dr[y == 1,0],c = 'black',label = iris.target_names[1])
plt.scatter(X_dr[y == 2,1],X_dr[y == 2,0],c = 'orange',label = iris.target_names[2])
plt.legend()
plt.title("PCA of IRIS dataset")

plt.show()
```

```python
colors = ["red","black","orange"]
plt.figure()
for i in range(3):
    plt.scatter(X_dr[y == i,0],
               X_dr[y == i,1],
                alpha = 0.7,
                c = colors[i],
                label = iris.target_names[i]
               )
    
plt.legend()
plt.title("PCA of IRIS datasets")
plt.show()
```

```python
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_ratio_.sum()
```

```python
import numpy as np
pca_line = PCA().fit(X)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4]) # 限制坐标轴显示整数
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance radio")
plt.show()
```

### 最大似然估计自选超参数

```python
# 输入‘mle’作为n_components的参数输入，就可以调用这种方法
pca_mel = PCA(n_components = 'mle')
x_mel = pca_mel.fit_transform(X)
x_mel
```

```python
pca_mel.explained_variance_
```

```python
pca_mel.explained_variance_ratio_.sum()
```

### 按信息量占比选超参数

```python
# 输入[0,1]之间的浮点数，并且让参数svd_solver ==‘full’,表示希望降维后的总解释性方差占比大于n_components
# 指定的百分比，即希望保留百分之多少的信息量
pca_f = PCA(n_components = 0.97,svd_solver = 'full')
x_f = pca_f.fit_transform(X)

x_f
```

```python
pca_f.explained_variance_ratio_
```
